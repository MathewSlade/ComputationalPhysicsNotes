```python
#I used the textbook to help me get started on this code and the equations
from numpy import arange
from pylab import plot, xlabel, ylabel, show

def Vin(t): #I got most of the code right and realized that I needed to make two separate functions. I looked up some examples.
    if ((2*t)//1)%2 == 0: #This is what I use to make sure I get alternating values of 1 and -1.
        return 1
    else: 
        return -1

def f(Vout,t):
    return (Vin(t)-Vout)/RC
a=0.0
b=10.0
N=500
h=(b-a)/N
RC=.01

tpoints = arange (a,b,h) #My x axis will be a range of values for t from a to b in steps that are "h" big.
#The second big bug in my code was that instead of tpoints I had t.
Voutpoints = [] #My y axis will be constructed based off of the t values, built piece by piece.
Vout=0.0

for t in tpoints:
    Voutpoints.append(Vout)
    k1=h*f(Vout,t)
    k2=h*f(Vout+(0.5*k1),t+(0.5*h))
    k3=h*f(Vout+(0.5*k2),t+(0.5*h))
    k4=h*f(Vout+k3,t+h)
    Vout += (k1+(2*k2)+(2*k3)+k4)/6
    
plot(tpoints,Voutpoints)
xlabel('t')
ylabel('Vout')
show()
```
